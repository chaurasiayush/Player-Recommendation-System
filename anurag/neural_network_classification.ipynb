```python
#Applying neural network for classification
import pandas
import numpy as np
from tensorflow_core.python.keras.models import Sequential
from tensorflow_core.python.keras.optimizers import SGD
from tensorflow_core.python.keras.layers import Dense,Dropout,Activation
from tensorflow_core.python.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow_core.python.keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
#from tensorflow_core.python.keras.utils import to_categorical
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from tensorflow.python.keras.utils.np_utils import to_categorical
```

```python
df = pandas.read_csv('Desktop/btech_project/clustered02.csv')
df.head(5)
```

```python
x1=df[['batRI','bowlRI']].values
y1=df['clusterRI'].values
#print(y)
```

```python
#modify()
x=preprocessing.scale(x1)
x=preprocessing.normalize(x)
y=to_categorical(y1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=Sequential()
model.add(Dense(20, input_dim=2, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(20, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(5, activation='softmax'))
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=2000,batch_size=128)
```

```python
score=model.evaluate(x_test,y_test,batch_size=128)
```
